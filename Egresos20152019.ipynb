```python
#importo la librería
import pandas as pd
```

```python
#Reporte: Egresos de los hospitales del Gobierno de la Ciudad de Buenos Aires por hospital, por mes. Ciudad de Buenos Aires. 2005/2019
#Guardo ubicacion de dataset y leo dataset: egresos por hospital, por mes. Datos: 2005 - 2017.

ubicacion_xlsx_prueba = 'https://www.estadisticaciudad.gob.ar/eyc/wp-content/uploads/2017/05/S_MVH_AX09.xlsx'
df_ma_internacion = pd.read_excel(ubicacion_xlsx_prueba, sheet_name='2017',
     header=1
)
#traigo el nombre de las columnas
df_ma_internacion.columns.values
```

```python
### PRUEBA ITERACION - FUNCIONA!
url_base = 'https://www.estadisticaciudad.gob.ar/eyc/wp-content/uploads/2017/05/S_MVH_AX09.xlsx'

dfs = []
# Loop para los años 2005 a 2017
for year in range(2005, 2018):
    # Leer el archivo Excel para el año actual
    df = pd.read_excel(url_base, sheet_name=str(year), header=1)
    # Agregar una columna con el año
    df['Año'] = year
    # Agregar el DataFrame a la lista
    dfs.append(df)

# Concatenar todos los DataFrames en uno solo
df_concatenado = pd.concat(dfs, ignore_index=True)

# Mostrar las primeras filas del DataFrame concatenado
print(df_concatenado.head())
```

```python
print(df_concatenado)
```
